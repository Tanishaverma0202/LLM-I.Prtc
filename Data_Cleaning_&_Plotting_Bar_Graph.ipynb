```python
import pandas as pd

#Create sample data
data = {
    "title": ["AI Revolution", "Python Basics", "Healthy Eating", "Travel Tips", "Machine Learning"],
    "author": ["Alice", "Bob", "Charlie", "David", "Eva"],
    "category": ["Tech", "Tech", "Health", "Travel", "Tech"],
    "views": [1500, 1200, 900, 600, 2000]
}

#Convert to DataFrame
df = pd.DataFrame(data)

#Save to CSV
df.to_csv("demo_articles.csv", index=False)

print("Demo CSV file created: demo_articles.csv")
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the data
df = pd.read_csv('demo_articles.csv')

#Inspect the data
print(df.head())
print(df.info())

# Clean the data by removing rows with nulls
df_clean = df.dropna(subset=['category'])  # remove rows where 'category' is null

#Count articles in each category
category_counts = df_clean['category'].value_counts()
print(category_counts)

# Plot a bar graph
plt.figure(figsize=(8,5))
category_counts.plot(kind='bar', color='skyblue')
plt.title('Number of Articles by Category')
plt.xlabel('Category')
plt.ylabel('Number of Articles')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
